#### Data Cleaning

```python
# Load dataset into a DataFrame, df
df = pd.read_csv("cardio_good_fitness.csv")

# Check data type of each column
df.info()

# Preview of first 5 rows
df.head()
```

##### Handling missing values

```python
# Sum up the total number of missing values in each column
df.isna().sum()
```

```python
# since the "age" column contains a mixture of data types due to the mixture of "na" in some columns,
# this is how the data type is changed from object to a float and how the "na" values are replaced.

# Convert Age column
df['Age'] = df['Age'].replace('Na', np.nan)  # Replace 'Na' with NaN
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')  # Convert to numeric
df['Age'] = df['Age'].fillna(df['Age'].median())  # Fill NaN with median
df['Age'] = df['Age'].astype(int)  # Convert to integer


# Results after conversion
print("Data type:", df['Age'].dtype)
```

```python
# Calculate the mean for "Age" column
mean_age = df['Age'].mean()
# Replace missing values in 'Age" column with the mean
df['Age'].fillna(mean_age, inplace=True)

# Calculate the mean for "Usage" column
mean_usage = df["Usage"].mean()
# Replace missing values in 'Usage" column with the mean
df['Usage'].fillna(mean_usage, inplace=True)

# Create a copy of the DataFrame
df_new = df.copy() 

print(df_new.isna().sum()) 
```

##### Checking for data consistency 

```python
# Check for negative values in the specified columns
negative_values_check = (df_new[['Age', 'Usage', 'Fitness', 'Income', 'Miles']] < 0).any()

# Print the result
print(negative_values_check)
```

```python
# Verify if all entries for the "Gender","Product","Branch","Fitness" and "MaritalStatus" columns

# 1. Verify Gender column ('Male' or 'Female')
valid_genders = ['Male', 'Female']
gender_valid = df_new['Gender'].isin(valid_genders)
invalid_gender_count = (gender_valid == False).sum()
print(f"Number of invalid Gender values: {invalid_gender_count}")

# 2. Verify Product column (valid models: TM195, TM498, TM798)
valid_products = ['TM195', 'TM498', 'TM798']
product_valid = df_new['Product'].isin(valid_products)
invalid_product_count = (product_valid == False).sum()
print(f"Number of invalid Product values: {invalid_product_count}")

# 3. Verify Branch column (valid branches: North, South, East, West)
valid_branches = ['North', 'South', 'East', 'West']
branch_valid = df_new['Branch'].isin(valid_branches)
invalid_branch_count = (branch_valid == False).sum()
print(f"Number of invalid Branch values: {invalid_branch_count}")

# 4. Verify MaritalStatus column ('Single' or 'Partnered')
valid_marital_status = ['Single', 'Partnered']
marital_status_valid = df_new['MaritalStatus'].isin(valid_marital_status)
invalid_marital_status_count = (marital_status_valid == False).sum()
print(f"Number of invalid MaritalStatus values: {invalid_marital_status_count}")
```

```python
# Since the Marital Status Columns have Columns with abbreviated versions of "Singled and "Partnered", we have to change them
# Create mapping dictionary
status_mapping = {
    'S': 'Single',
    'P': 'Partnered'
}

# Replace values using map
df['MaritalStatus'] = df['MaritalStatus'].map(status_mapping).fillna(df['MaritalStatus'])

# Print value counts to verify changes
print("\nMarital Status Distribution After Cleaning:")
# Check for any remaining unexpected values
unique_values = df['MaritalStatus'].unique()
print("\nUnique values in marital status column:", unique_values)
```

```python
# 1. What is the distribution of sales across different treadmill models? 
# Count sales by product
product_sales = df['Product'].value_counts()

# Set colors
colors = ['#FF9999', '#66B2FF', '#99FF99']

# Create figure
plt.figure(figsize=(12, 8))

# Create pie chart with customizations
wedges, texts, autotexts = plt.pie(product_sales, 
                                  labels=product_sales.index,
                                  autopct='%1.1f%%',
                                  startangle=90,
                                  colors=colors,
                                  explode=[0.05] * len(product_sales))

# Enhance text properties
plt.setp(autotexts, size=10, weight="bold", color='black')
plt.setp(texts, size=12)

# Add title
plt.title('Distribution of Treadmill Sales by Model\n', 
          fontsize=12, 
          fontweight='bold')

# Add legend
plt.legend(title="Treadmill Models",
          title_fontsize=12,
          fontsize=10,
          bbox_to_anchor=(1.2, 0.5),
          loc='center')

# Add sales volume
total_sales = product_sales.sum()
plt.annotate(f'Total Sales: {total_sales}',
            xy=(1.2, -0.1),
            xycoords='axes fraction',
            fontsize=10)

plt.tight_layout()
plt.show()

# print summary statistics
print("\nSales Distribution across Models:")
sales_dist = df['Product'].value_counts()
sales_percent = df['Product'].value_counts(normalize=True) * 100
summary_df = pd.DataFrame({
    'Count': sales_dist,
    'Percentage': sales_percent.round(1)
})
print(summary_df)
```

### Visualisation 2

```python
# 2. Is there a relationship between customer fitness level and product choice?

# Create cross-tabulation
fitness_counts = pd.crosstab(df['Fitness'], df['Product'])
fitness_pcts = pd.crosstab(df['Fitness'], df['Product'], normalize='columns') * 100

# Create bar chart
fig, ax = plt.subplots(figsize=(12, 6))

# Set width and positions
x = np.arange(len(fitness_counts.index))
width = 0.25

# Plot bars for each product
for i, product in enumerate(fitness_counts.columns):
    data = fitness_pcts[product]
    bars = ax.bar(x + i*width, data, width, label=product,
                  color=['#FF9999', '#66B2FF', '#99FF99'][i])
    
    # Add value labels
    for bar in bars:
        height = bar.get_height()
        if height > 0:
            ax.text(bar.get_x() + bar.get_width()/2,
                   height,
                   f'{height:.1f}%',
                   ha='center', va='bottom')

# Customize plot
plt.title('Distribution of Fitness Levels by Treadmill Model', pad=20, fontsize=14, fontweight='bold')
plt.xlabel('Fitness Level', fontsize=12)
plt.ylabel('Percentage (%)', fontsize=12)
plt.legend(title='Treadmill Model')
plt.grid(axis='y', linestyle='--', alpha=0.3)

# Set x-axis labels
ax.set_xticks(x + width)
ax.set_xticklabels([f'Level {i}' for i in range(1, 6)])

plt.tight_layout()
plt.show()

print("\nFitness Level Distribution by Product (%):")
print(fitness_pcts.round(1))

print("\nCount of Customers by Fitness Level and Product:")
print(fitness_counts)
```

```python
# 4. How does education level influence product choice
# Calculate distribution of original education levels
edu_product = pd.crosstab(df['Product'], df['Education'])

# Set up the figure
plt.figure(figsize=(15, 7))

# Set width and positions of bars
width = 0.1  # narrower bars since we have more categories
x = np.arange(len(edu_product.index))

colors = ['#4BA3C3', '#175676', '#BAE4E5', '#63C7B2', '#449DD1','#8DAA91', '#4F7942', '#679267']

# Create bars for each education level
for i, (level, color) in enumerate(zip(sorted(edu_product.columns), colors)):
    position = x + (i - len(edu_product.columns)/2 + 0.5) * width
    plt.bar(position, edu_product[level], width, 
            label=f'{level} years', color=color, alpha=0.8)
    
    # Add value labels on bars
    for j, value in enumerate(edu_product[level]):
        if value > 0:  # Only add label if there's a value
            plt.text(position[j], value, str(value), 
                    ha='center', va='bottom', fontsize=9)

# Customize plot
plt.title('Product Choice by Years of Education', fontsize=14, pad=20)
plt.xlabel('Product Model', fontsize=12)
plt.ylabel('Number of Customers', fontsize=12)

# Set x-axis ticks
plt.xticks(x, edu_product.index)

# Add legend
plt.legend(title='Education (Years)', bbox_to_anchor=(1.05, 1), loc='upper left')

# Add grid
plt.grid(True, axis='y', linestyle='--', alpha=0.3)

# Remove top and right spines
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Print insights
print("Key Insights:")
print("\nTotal customers by years of education:")
print(edu_product.sum())

print("\nProduct preferences by education level (%):")
percentages = edu_product.div(edu_product.sum(axis=1), axis=0) * 100
print(percentages.round(1))

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

```python
# 7. What are the top 2 age groups that represent the primary customer base?
def plot_age_distribution(df):
    # Create age groups
    df['age_group'] = pd.cut(df['Age'], 
                            bins=[17, 20, 25, 30, 35, 50],
                            labels=['18-20', '21-25', '26-30', '31-35', '36-50'])
    
    # Calculate distribution
    age_dist = df['age_group'].value_counts().sort_index()
    age_dist_pct = (age_dist / len(df) * 100).round(1)
    
    # Create figure with two y-axes for histogram and density line
    fig, ax1 = plt.subplots(figsize=(12, 6))
    
    # Create histogram bars
    colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEEAD']
    bars = ax1.bar(range(len(age_dist)), age_dist.values, 
                   color=colors[:len(age_dist)], alpha=0.6)
    
    # Create smooth line
    x = np.array(range(len(age_dist)))
    y = age_dist.values
    
    # Create smooth curve using spline interpolation
    X_Y_Spline = make_interp_spline(x, y)
    X_ = np.linspace(x.min(), x.max(), 500)
    Y_ = X_Y_Spline(X_)
    
    # Scale the smoothed line to match bar heights
    max_height = max(age_dist.values)
    Y_ = Y_ * max_height / max(Y_)
    
    # Plot the smoothed line
    ax1.plot(X_, Y_, color='#FF1493', linewidth=2)
    
    # Add value labels on top of the bars
    for i, (count, pct) in enumerate(zip(age_dist.values, age_dist_pct.values)):
        ax1.text(i, count, f'{count}\n({pct:.1f}%)', 
                 ha='center', va='bottom', fontsize=10)
    
    # Customize the plot
    plt.title('Age Distribution of Customers', fontsize=14, pad=20)
    plt.xlabel('Age Group', fontsize=12)
    ax1.set_ylabel('Number of Customers', fontsize=12)
    
    # Set x-ticks
    plt.xticks(range(len(age_dist)), age_dist.index, rotation=0)
    
    # Add grid
    ax1.grid(True, axis='y', linestyle='--', alpha=0.3)
    
    # Remove top and right spines
    ax1.spines['top'].set_visible(False)
    ax1.spines['right'].set_visible(False)
    
    # Adjust layout
    plt.tight_layout()
    
    # Print detailed distribution
    print("\nDetailed Age Group Distribution (Sorted by Count):")
    sorted_dist = age_dist.sort_values(ascending=False)
    sorted_pct = age_dist_pct[sorted_dist.index]
    for age_group, count in sorted_dist.items():
        percentage = sorted_pct[age_group]
        print(f"{age_group}: {count} customers ({percentage:.1f}% of total)")
    print(f"\nTotal Customers: {age_dist.sum()}")
    
    return fig

fig = plot_age_distribution(df)
plt.show()
```

```python
# 14. What are the top 2 age groups that contribute to the highest sales each product?
def plot_age_product_distribution(df):
    # Create age groups if not already present
    if 'age_group' not in df.columns:
        df['age_group'] = pd.cut(df['Age'],
                                bins=[17, 20, 25, 30, 35, 50],
                                labels=['18-20', '21-25', '26-30', '31-35', '36-50'])

    # Calculate age group counts for each product
    age_product_counts = pd.crosstab(df['Product'], df['age_group'])

    # Create the figure
    fig, ax = plt.subplots(figsize=(12, 6))

    # Define colors for age groups
    colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEEAD']
    
    # Initialize bottom values for stacked bars
    bottom_values = np.zeros(len(age_product_counts))

    # Plot each age group
    for i, age_group in enumerate(age_product_counts.columns):
        counts = age_product_counts[age_group]
        bars = ax.bar(range(len(age_product_counts)), counts,
                     bottom=bottom_values,
                     color=colors[i],
                     label=age_group)

        # Add labels for each segment
        for j in range(len(counts)):
            count = counts.iloc[j]
            if count > 0:
                # Position the text in the middle of each segment
                ax.text(j, bottom_values[j] + count/2,
                       str(count),
                       ha='center', va='center',
                       fontsize=10)

        # Update bottom values for next layer
        bottom_values += counts.values

    # Customize the plot
    plt.title('Age Group Distribution by Product Model', fontsize=14, pad=20)
    plt.xlabel('Product Model', fontsize=12)
    plt.ylabel('Number of Customers', fontsize=12)

    # Set x-axis ticks
    plt.xticks(range(len(age_product_counts)), age_product_counts.index)

    # Add legend
    plt.legend(title='Age Group', bbox_to_anchor=(1.05, 1), loc='upper left')

    # Add grid
    plt.grid(True, axis='y', linestyle='--', alpha=0.3)

    # Remove top and right spines
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    # Print summary statistics
    print("\nAge Group Distribution by Product:")
    print(age_product_counts)

    print("\nPercentage Distribution:")
    percentages = (age_product_counts.div(age_product_counts.sum(axis=1), axis=0) * 100).round(1)
    print(percentages)

    # Find top 2 age groups for each product
    print("\nTop 2 Age Groups by Product:")
    for product in age_product_counts.index:
        product_dist = age_product_counts.loc[product]
        top_2_age_groups = product_dist.nlargest(2)
        print(f"\n{product}:")
        for age_group, count in top_2_age_groups.items():
            percentage = percentages.loc[product, age_group]
            print(f"{age_group}: {count} customers ({percentage}%)")

    # Adjust layout
    plt.tight_layout()

    return fig
fig = plot_age_product_distribution(df)
plt.show()
```

```python
# Dashboard 3: Usage Patterns Dashboard
def create_usage_dashboard(df):
    # Create main figure
    fig = plt.figure(figsize=(20, 10))
    gs = fig.add_gridspec(1, 2, width_ratios=[1, 1], wspace=0.2)
    
    # 1. Usage by Fitness Level (Left)
    ax1 = fig.add_subplot(gs[0])
    
    # Get data for each fitness level
    fitness_levels = sorted(df['Fitness'].unique())
    data = [df[df['Fitness'] == level]['Usage'] for level in fitness_levels]
    
    # Create boxplot
    colors = ['#A8E6CF', '#DCEDC1', '#FFD3B6', '#FFAAA5', '#FF8B94']
    bp = ax1.boxplot(data, 
                     patch_artist=True,
                     medianprops=dict(color="black", linewidth=1.5),
                     flierprops=dict(marker='o', markerfacecolor='black', markersize=5),
                     whiskerprops=dict(color='black'),
                     capprops=dict(color='black'))
    
    # Color each box
    for patch, color in zip(bp['boxes'], colors):
        patch.set_facecolor(color)
        patch.set_alpha(0.7)
    
    # Customize first plot
    ax1.set_title('Distribution of Treadmill Usage by Fitness Level', 
                  fontsize=14, pad=20)
    ax1.set_xlabel('Fitness Level', fontsize=12)
    ax1.set_ylabel('Usage (hours/week)', fontsize=12)
    ax1.set_xticklabels(fitness_levels)
    ax1.grid(True, axis='y', linestyle='--', alpha=0.3)
    ax1.spines['top'].set_visible(False)
    ax1.spines['right'].set_visible(False)
    
    # 2. Usage by Age Group (Right)
    ax2 = fig.add_subplot(gs[1])
    
    # Create age groups
    df['age_group'] = pd.cut(df['Age'], 
                            bins=[0, 20, 25, 30, 35, 50],
                            labels=['18-20', '21-25', '26-30', '31-35', '36-50'])
    
    # Create violin plot with box plot inside
    sns.violinplot(data=df, x='age_group', y='Usage',
                   palette=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEEAD'],
                   inner='box', alpha=0.7,
                   ax=ax2)
    
    # Add scatter plot
    for i, age_group in enumerate(df['age_group'].unique()):
        group_data = df[df['age_group'] == age_group]
        x = np.random.normal(i, 0.05, size=len(group_data))
        ax2.scatter(x, group_data['Usage'],
                   alpha=0.4,
                   s=30,
                   color='black',
                   zorder=2)
    
    # Customize second plot
    ax2.set_title('Weekly Usage Distribution by Age Group', 
                  fontsize=14, pad=20)
    ax2.set_xlabel('Age Group', fontsize=12)
    ax2.set_ylabel('Weekly Usage (hours)', fontsize=12)
    ax2.grid(True, axis='y', linestyle='--', alpha=0.3)
    ax2.spines['top'].set_visible(False)
    ax2.spines['right'].set_visible(False)
    
    # Add correlation information
    correlation = df['Age'].corr(df['Usage'])
    ax2.text(0.95, 0.95, f'Age-Usage Correlation: {correlation:.3f}',
             transform=ax2.transAxes,
             ha='right', va='top',
             bbox=dict(facecolor='white', alpha=0.8, edgecolor='none'))
    
    # Add main title
    fig.suptitle('Treadmill Usage Analysis by Fitness Level and Age Group', 
                 fontsize=16, fontweight='bold', y=0.95)
    
    # Adjust layout
    plt.tight_layout()
    
    return fig

# Create and display dashboard
fig = create_usage_dashboard(df)
plt.show()

# Print summary statistics
print("\nSummary Statistics by Fitness Level:")
fitness_stats = df.groupby('Fitness')['Usage'].agg(['count', 'mean', 'median', 'std']).round(2)
print(fitness_stats)

print("\nSummary Statistics by Age Group:")
age_stats = df.groupby('age_group', observed=True)['Usage'].agg(['count', 'mean', 'median', 'std']).round(2)
print(age_stats)
```

```python
# Dashboard 4: Geographic Analysis Dashboard
def create_branch_dashboard(df):
   # Create main figure
   fig = plt.figure(figsize=(20, 10))
   gs = fig.add_gridspec(1, 2, width_ratios=[1.2, 1], wspace=0.2)
   
   # 1. Sales Distribution Heatmap (Left)
   ax1 = fig.add_subplot(gs[0])
   
   # Create cross-tabulation of data
   sales_data = pd.crosstab(df['Branch'], df['Product'])
   percentage_dist = (sales_data.div(sales_data.sum(axis=1), axis=0) * 100).round(1)
   
   # Create heatmap
   heatmap = sns.heatmap(sales_data, 
                        cmap='RdPu',
                        annot=False,
                        fmt='d',
                        cbar_kws={'label': 'Number of Sales'},
                        cbar=True,
                        linewidths=0.5,
                        linecolor='white',
                        ax=ax1)
   
   # Add custom annotations
   for i in range(len(sales_data.index)):
       for j in range(len(sales_data.columns)):
           count = sales_data.iloc[i, j]
           percentage = percentage_dist.iloc[i, j]
           text = f'{count}\n({percentage:.1f}%)'
           
           heatmap.text(j + 0.5, i + 0.5, text,
                       ha='center', va='center',
                       fontsize=12,
                       fontweight='bold',
                       color='black' if percentage < 50 else 'white')
   
   # Customize first plot
   ax1.set_title('Sales Distribution Across Branches', 
                 fontsize=14, pad=20)
   ax1.set_xlabel('Product Model', fontsize=12)
   ax1.set_ylabel('Branch', fontsize=12)
   ax1.set_xticklabels(ax1.get_xticklabels(), rotation=0, fontsize=12)
   ax1.set_yticklabels(ax1.get_yticklabels(), rotation=0, fontsize=12)
   
   # 2. Usage by Branch (Right)
   ax2 = fig.add_subplot(gs[1])
   
   # Create box plot
   sns.boxplot(data=df, x='Branch', y='Usage',
               palette=['#FEE1E8', '#FEB4CB','#FF367C', '#D10063'],
               width=0.7,
               ax=ax2)
   
   # Add mean values
   means = df.groupby('Branch', observed=True)['Usage'].mean()
   for i, mean in enumerate(means):
       ax2.text(i, mean, f'Mean: {mean:.1f}h',
               ha='center', va='bottom',
               fontweight='bold')
   
   # Customize second plot
   ax2.set_title('Average Weekly Usage by Branch', 
                 fontsize=14, pad=20)
   ax2.set_xlabel('Branch', fontsize=12)
   ax2.set_ylabel('Weekly Usage (hours)', fontsize=12)
   ax2.grid(True, axis='y', linestyle='--', alpha=0.3)
   ax2.spines['top'].set_visible(False)
   ax2.spines['right'].set_visible(False)
   
   # Add main title
   fig.suptitle('Branch Performance Analysis', 
                fontsize=16, fontweight='bold', y=0.95)
   
   # Adjust layout
   plt.tight_layout()
   
   return fig

# Create and display dashboard
fig = create_branch_dashboard(df)
plt.show()

# Print summary statistics
print("\nSales Distribution by Branch:")
sales_data = pd.crosstab(df['Branch'], df['Product'])
print(sales_data)

print("\nBranch Usage Statistics:")
stats = df.groupby('Branch', observed=True)['Usage'].agg(['count', 'mean', 'median', 'std']).round(2)
print(stats)

max_branch = stats['mean'].idxmax()
max_mean = stats.loc[max_branch, 'mean']
print(f"\nHighest Average Usage:")
print(f"{max_branch}: {max_mean:.2f} hours per week")
```

```python
# Dashboard 5. Price Sensitivity Dashboard
def create_sales_demographics_dashboard(df):
   # Create main figure with custom layout
   fig = plt.figure(figsize=(20, 12))
   gs = fig.add_gridspec(2, 3, height_ratios=[1, 1.2], width_ratios=[1, 1, 1], hspace=0.4)
   
   # 1. Income Distribution by Product (Top Row)
   products = df['Product'].unique()
   colors = ['#FF9999', '#66B2FF', '#99FF99']
   
   for i, product in enumerate(products):
       ax = fig.add_subplot(gs[0, i])
       
       # Filter data for specific product
       product_data = df[df['Product'] == product]['Income']
       
       # Plot histogram with density
       ax.hist(product_data, bins='auto', density=True, alpha=0.7, 
               edgecolor='black', color=colors[i])
       
       # Add kernel density estimate
       kde = gaussian_kde(product_data)
       x_range = np.linspace(product_data.min(), product_data.max(), 100)
       ax.plot(x_range, kde(x_range), '-', color=colors[i], linewidth=2)
       
       ax.set_title(f'Income Distribution for {product}', fontsize=12)
       ax.set_xlabel('Income ($)', fontsize=10)
       if i == 0:
           ax.set_ylabel('Density', fontsize=10)
           
   # 2. Age Group Distribution (Bottom Row)
   ax2 = fig.add_subplot(gs[1, :])
   
   # Create age groups
   if 'age_group' not in df.columns:
       df['age_group'] = pd.cut(df['Age'],
                               bins=[17, 20, 25, 30, 35, 50],
                               labels=['18-20', '21-25', '26-30', '31-35', '36-50'])
   
   # Calculate age group counts
   age_product_counts = pd.crosstab(df['Product'], df['age_group'])
   
   # Define colors for age groups
   age_colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEEAD']
   
   # Initialize bottom values for stacked bars
   bottom_values = np.zeros(len(age_product_counts))
   
   # Plot stacked bars
   for i, age_group in enumerate(age_product_counts.columns):
       counts = age_product_counts[age_group]
       bars = ax2.bar(range(len(age_product_counts)), counts,
                     bottom=bottom_values,
                     color=age_colors[i],
                     label=age_group)
       
       # Add labels
       for j in range(len(counts)):
           count = counts.iloc[j]
           if count > 0:
               ax2.text(j, bottom_values[j] + count/2,
                       str(count),
                       ha='center', va='center',
                       fontsize=10)
       bottom_values += counts.values
   
   # Customize bottom plot
   ax2.set_title('Age Group Distribution by Product Model', fontsize=14, pad=20)
   ax2.set_xlabel('Product Model', fontsize=12)
   ax2.set_ylabel('Number of Customers', fontsize=12)
   ax2.set_xticks(range(len(age_product_counts)))
   ax2.set_xticklabels(age_product_counts.index)
   ax2.legend(title='Age Group', bbox_to_anchor=(1.05, 1))
   ax2.grid(True, axis='y', linestyle='--', alpha=0.3)
   ax2.spines['top'].set_visible(False)
   ax2.spines['right'].set_visible(False)
   
   # Add main title
   fig.suptitle('Product Demographics Analysis: Income and Age Distribution', 
                fontsize=16, fontweight='bold', y=0.95)
   
   return fig

# Create and display dashboard
fig = create_sales_demographics_dashboard(df)
plt.show()
```
